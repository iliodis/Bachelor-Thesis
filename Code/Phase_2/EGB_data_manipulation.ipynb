```python
from pathlib import Path
import glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# from mpl_toolkits.axes_grid1 import make_axes_locatable
import matplotlib as mpl
```

```python
path0 = Path(r"C:\Users\Desktop\Giannis\EGB\MR_EGB-main\MR_data")
```

```python
file_list = glob.glob(str(path0.joinpath("*")))
```

```python
# Units
units = {
    'M'  :'M_{\odot}',
    'R'  :'km',
    'p_c':'10^{25} dyn/cm^2'
}
```

```python
df = pd.DataFrame({})
alpha_list = []
for file_name in file_list:
    
    alpha = file_name[65:]
    temp  = pd.read_table(file_name, sep="\s+", names= ['M','R','p_c'])
    
    temp['a'] = np.ones(len(temp))*float(alpha)
    
    alpha_list.append(float(alpha))
    
    frames = [df,temp]
    df     = pd.concat(frames, ignore_index=True)
    
```

```python
# All linear
n_lines = int(len(df)/200)

a = alpha_list
clrs = a.copy()
clrs.sort()

norm = mpl.colors.Normalize(vmin=min(a), vmax=max(a))
cmap = mpl.cm.ScalarMappable(norm=norm, cmap=mpl.cm.coolwarm)
cmap.set_array([])

fig, ax = plt.subplots(1,3, figsize = (18,6))
for i in range(len(alpha_list)):
    temp = df[df['a'] == alpha_list[i]]
    
    r  = temp.R.values
    m  = temp.M.values
    pc = temp.p_c.values
    
    ax[0].plot(r, m, c=cmap.to_rgba(alpha_list[i]))
    # ax[0].tick_params(axis='both', which='major', labelsize=10)
    
    ax[1].plot(pc, m, c=cmap.to_rgba(alpha_list[i]))
    
    ax[2].plot(pc, r, c=cmap.to_rgba(alpha_list[i]))


cbar = fig.colorbar(cmap, ticks=clrs[::3])
cbar.set_label('$a$ [$km^2$]', fontsize = 20, rotation=90)

name = "$R$ [$"+units["R"]+"$]"
ax[0].set_xlabel(name, fontsize = 20)
name = "$M$ [$"+units["M"]+"$]"
ax[0].set_ylabel(name, fontsize = 20)

name = "$p_c$ [$"+units["p_c"]+"$]"
ax[1].set_xlabel(name, fontsize = 20)
name = "$M$ [$"+units["M"]+"$]"
ax[1].set_ylabel(name, fontsize = 20)

name = "$p_c$ [$"+units["p_c"]+"$]"
ax[2].set_xlabel(name, fontsize = 20)
name = "$R$ [$"+units["R"]+"$]"
ax[2].set_ylabel(name, fontsize = 20)

plt.show()




# All log
n_lines = int(len(df)/200)

a = alpha_list
clrs = a.copy()
clrs.sort()

norm = mpl.colors.Normalize(vmin=min(a), vmax=max(a))
cmap = mpl.cm.ScalarMappable(norm=norm, cmap=mpl.cm.coolwarm)
cmap.set_array([])

fig, ax = plt.subplots(1,3, figsize = (18,6))
for i in range(len(alpha_list)):
    temp = df[df['a'] == alpha_list[i]]
    
    r  = temp.R.values
    m  = temp.M.values
    pc = temp.p_c.values
    
    ax[0].loglog(r, m, c=cmap.to_rgba(alpha_list[i]))
    # ax[0].tick_params(axis='both', which='major', labelsize=10)
    
    ax[1].loglog(pc, m, c=cmap.to_rgba(alpha_list[i]))
    
    ax[2].loglog(pc, r, c=cmap.to_rgba(alpha_list[i]))


cbar = fig.colorbar(cmap, ticks=clrs[::3])
cbar.set_label('$a$ [$km^2$]', fontsize = 20, rotation=90)

name = "$R$ [$"+units["R"]+"$]"
ax[0].set_xlabel(name, fontsize = 20)
name = "$M$ [$"+units["M"]+"$]"
ax[0].set_ylabel(name, fontsize = 20)

name = "$p_c$ [$"+units["p_c"]+"$]"
ax[1].set_xlabel(name, fontsize = 20)
name = "$M$ [$"+units["M"]+"$]"
ax[1].set_ylabel(name, fontsize = 20)

name = "$p_c$ [$"+units["p_c"]+"$]"
ax[2].set_xlabel(name, fontsize = 20)
name = "$R$ [$"+units["R"]+"$]"
ax[2].set_ylabel(name, fontsize = 20)

plt.show()




# p_c log
n_lines = int(len(df)/200)

a = alpha_list
clrs = a.copy()
clrs.sort()

norm = mpl.colors.Normalize(vmin=min(a), vmax=max(a))
cmap = mpl.cm.ScalarMappable(norm=norm, cmap=mpl.cm.coolwarm)
cmap.set_array([])

fig, ax = plt.subplots(1,3, figsize = (18,6))
for i in range(len(alpha_list)):
    temp = df[df['a'] == alpha_list[i]]
    
    r  = temp.R.values
    m  = temp.M.values
    pc = temp.p_c.values
    
    ax[0].plot(r, m, c=cmap.to_rgba(alpha_list[i]))
    # ax[0].tick_params(axis='both', which='major', labelsize=10)
    
    ax[1].semilogx(pc, m, c=cmap.to_rgba(alpha_list[i]))
    
    ax[2].semilogx(pc, r, c=cmap.to_rgba(alpha_list[i]))


cbar = fig.colorbar(cmap, ticks=clrs[::3])
cbar.set_label('$a$ [$km^2$]', fontsize = 20, rotation=90)

name = "$R$ [$"+units["R"]+"$]"
ax[0].set_xlabel(name, fontsize = 20)
name = "$M$ [$"+units["M"]+"$]"
ax[0].set_ylabel(name, fontsize = 20)

name = "$p_c$ [$"+units["p_c"]+"$]"
ax[1].set_xlabel(name, fontsize = 20)
name = "$M$ [$"+units["M"]+"$]"
ax[1].set_ylabel(name, fontsize = 20)

name = "$p_c$ [$"+units["p_c"]+"$]"
ax[2].set_xlabel(name, fontsize = 20)
name = "$R$ [$"+units["R"]+"$]"
ax[2].set_ylabel(name, fontsize = 20)

plt.show()



# p_c log R log
n_lines = int(len(df)/200)

a = alpha_list
clrs = a.copy()
clrs.sort()

norm = mpl.colors.Normalize(vmin=min(a), vmax=max(a))
cmap = mpl.cm.ScalarMappable(norm=norm, cmap=mpl.cm.coolwarm)
cmap.set_array([])

fig, ax = plt.subplots(1,3, figsize = (18,6))
for i in range(len(alpha_list)):
    temp = df[df['a'] == alpha_list[i]]
    
    r  = temp.R.values
    m  = temp.M.values
    pc = temp.p_c.values
    
    ax[0].semilogx(r, m, c=cmap.to_rgba(alpha_list[i]))
    # ax[0].tick_params(axis='both', which='major', labelsize=10)
    
    ax[1].semilogx(pc, m, c=cmap.to_rgba(alpha_list[i]))
    
    ax[2].loglog(pc, r, c=cmap.to_rgba(alpha_list[i]))


cbar = fig.colorbar(cmap, ticks=clrs[::3])
cbar.set_label('$a$ [$km^2$]', fontsize = 20, rotation=90)

name = "$R$ [$"+units["R"]+"$]"
ax[0].set_xlabel(name, fontsize = 20)
name = "$M$ [$"+units["M"]+"$]"
ax[0].set_ylabel(name, fontsize = 20)

name = "$p_c$ [$"+units["p_c"]+"$]"
ax[1].set_xlabel(name, fontsize = 20)
name = "$M$ [$"+units["M"]+"$]"
ax[1].set_ylabel(name, fontsize = 20)

name = "$p_c$ [$"+units["p_c"]+"$]"
ax[2].set_xlabel(name, fontsize = 20)
name = "$R$ [$"+units["R"]+"$]"
ax[2].set_ylabel(name, fontsize = 20)

plt.show()




# p_c log M log
n_lines = int(len(df)/200)

a = alpha_list
clrs = a.copy()
clrs.sort()

norm = mpl.colors.Normalize(vmin=min(a), vmax=max(a))
cmap = mpl.cm.ScalarMappable(norm=norm, cmap=mpl.cm.coolwarm)
cmap.set_array([])

fig, ax = plt.subplots(1,3, figsize = (18,6))
for i in range(len(alpha_list)):
    temp = df[df['a'] == alpha_list[i]]
    
    r  = temp.R.values
    m  = temp.M.values
    pc = temp.p_c.values
    
    ax[0].semilogy(r, m, c=cmap.to_rgba(alpha_list[i]))
    # ax[0].tick_params(axis='both', which='major', labelsize=10)
    
    ax[1].loglog(pc, m, c=cmap.to_rgba(alpha_list[i]))
    
    ax[2].semilogx(pc, r, c=cmap.to_rgba(alpha_list[i]))


cbar = fig.colorbar(cmap, ticks=clrs[::3])
cbar.set_label('$a$ [$km^2$]', fontsize = 20, rotation=90)

name = "$R$ [$"+units["R"]+"$]"
ax[0].set_xlabel(name, fontsize = 20)
name = "$M$ [$"+units["M"]+"$]"
ax[0].set_ylabel(name, fontsize = 20)

name = "$p_c$ [$"+units["p_c"]+"$]"
ax[1].set_xlabel(name, fontsize = 20)
name = "$M$ [$"+units["M"]+"$]"
ax[1].set_ylabel(name, fontsize = 20)

name = "$p_c$ [$"+units["p_c"]+"$]"
ax[2].set_xlabel(name, fontsize = 20)
name = "$R$ [$"+units["R"]+"$]"
ax[2].set_ylabel(name, fontsize = 20)

plt.show()




```
